```python
#Excel檔轉入Py轉清單
import pandas 
data = []
with open('PPA 網址.xlsx', 'rb') as f:
    pd = pandas.read_excel(f)
for a in pd["課程網址"]:
    data.append(a)
data
```

```python
from selenium import webdriver
#網路驅動程序
from selenium.webdriver.chrome.options import Options
#Chrome選項
from selenium.webdriver.support.wait import WebDriverWait
#網頁緩衝器
from selenium.webdriver.support import expected_conditions as ec
#條件設定
from selenium.webdriver.common.by import By
#基本設置
from selenium.webdriver.common.keys import Keys
#給按鈕
from selenium.common.exceptions import NoSuchElementException as NEE
#判斷空值
from time import sleep
import json
import openpyxl as opx
#編輯exl


wb = opx.Workbook()
ws = wb.active

ws["A1"] = "課程名稱"
ws["B1"] = "特惠價錢"
ws["C1"] = "原價錢"
ws["D1"] = "目前學員人數"
ws["E1"] = "課程種類"
ws["F1"] = "老師名稱"

option = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values":{"notifications":2}}
option.add_experimental_option("prefs",prefs)

driver = webdriver.Chrome("C:\\Users\\htuvm\\python\\chrome動態爬蟲工具\\chromedriver.exe",options=option)


frequency = 1
for url in data:
    driver.get(url)
    f1 = open("18解鎖.json")
    cookies = json.loads(f1.read())
    for x in cookies:
        driver.add_cookie(x)
    f1.close
    driver.refresh()
    sleep(2)
    #1名稱
    try:
        name = (WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CLASS_NAME,"project-info-title"))).text)
    except:
        name = "無"
    #2特惠價格
    DDprice = (WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.XPATH,'/html/body/div[1]/div[3]/div[2]/div/div[2]/div[1]/div[1]/div/div[2]/div/div[2]/div[1]/div[1]'))).text)
    int_DDprice= ""
    for num in DDprice:
        if num.isdigit():
            int_DDprice+=num
    int_DDprice = int(int_DDprice)
    int_DDprice = float(int_DDprice)
    #原價
    try:
        price = (WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CLASS_NAME,"desc"))).text)
        int_price= ""
        for num1 in price:
            if num1.isdigit():
                int_price+=num1
        int_price = int(int_price)
        int_price = float(int_price)
    except:
        int_price = 0
    if int_price == 0:
        int_price = int_DDprice
    #3學員
    try:
        students = (WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CLASS_NAME,'project-info-meta-item'))).text)
        newS = ""
        for a in students:
            if a.isdigit() or a == "." or a == "K":
                newS+=a
        if "K" in newS:
            newS = float(newS[:newS.find("K")])*1000
        else:
            newS = float(newS)
    except:
        newS = 0
    if int_DDprice == newS:
        int_DDprice = 0
        newS = 0
    #4種類
    type_ = (WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CLASS_NAME,'project-label'))).text)          
    #5老師
    teacher = (WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CLASS_NAME,'project-info-creator-name'))).text)
    #銷售狀態
    try:
        mode = (WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CLASS_NAME,'icon icon-time_calendar_M'))).text) 
        mode = "預售中"
    except:
        mode ="已開課"
    ws.append([name,int_DDprice,int_price,newS,type_,teacher,mode])
    wb.save("爬蟲專題2.xlsx")
    print("第",frequency,"個")
    frequency+=1
    print("名稱",name)
    print("特惠價錢",int_DDprice)
    print("原價",int_price)
    print("學員",newS)
    print("種類",type_)
    print("作者",teacher)
    print("模式",mode)
    sleep(3)
#拉出所有課程網址爬出目標資料   
```
