```python
import jieba
import openpyxl as opx
wb = opx.Workbook()
ws = wb.active
ws["A1"]="詞彙"
ws["B1"]="數量"
with open("文字雲.csv","r",encoding="utf-8") as f:
    excel=f.read()
# print(excel)
a=jieba.lcut(excel)#精確模式
# print(a)
# print(a)
import operator
dic={}
for ele in a:
    if ele in dic:
        dic[ele]+=1
    else:
        dic[ele]=1
# print(dic)

sort=sorted(dic.items(),key=operator.itemgetter(1),reverse=True)
print(sort)

for ele in sort:
    print(ele[0],ele[1])
    if len(ele[0])>1:
        x=[]
        x.append(ele[0])
        x.append(ele[1])
        ws.append(x)

wb.save("course.xlsx")
```
