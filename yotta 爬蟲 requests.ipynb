```python
import requests
from bs4 import BeautifulSoup
import json
from time import sleep
import openpyxl as opx
import datetime

wb = opx.Workbook()
ws = wb.active
ws["A1"] = "課程名稱"
ws["B1"] = "優惠價格"
ws["C1"] = "課程原價"
ws["D1"] = "學員人數"
ws["E1"] = "作者名稱"
ws["F1"] = "課程類別"
ws["G1"] = "課程時數"
ws["I1"] = "課程狀態"
xx = 67
type_ = {"67":"商業","68":"設計","69":"科技","70":"語言","71":"生活"}
#編號轉文字
page = {"商業":21,"設計":27,"科技":4,"語言":6,"生活":30}
#文字轉頁數
while xx<72:
    Stype = type_[str(xx)]
    print(Stype)
    for x in range(0,page[Stype]):
        POST = {'type':str(xx),                      
        'subType':'all',
        'order':'latest',
        'courseType':'all',
        'groupID':'0',
        'page':str(x),
        'key':None,
        'force': '0'}

        url = "https://www.yottau.com.tw/explore/more"
        data = requests.post(url,POST)
        dj = data.json()
        print("第",x+1,"頁")
        print("**********************************************************************************************************************")
        for Intranet in dj["data"]["result"]:
            name = Intranet["course_name"]
            DDprice = Intranet["course_price"]
            price = Intranet["course_price_fake"]
            if price == 0:
                price = DDprice
            students =Intranet["course_buyer_current"]
            teacher = Intranet["user"]["name"]
            time_H =Intranet["online"]["course_duration"]["hours"]
            time_M =Intranet["online"]["course_duration"]["mins"]
            Tyep = Intranet["course_backer_end"]
            if Tyep == None:
                Tyep = "已開課"
            else:
                Tyep = "預購中"
            print('名稱',name)
            print('優惠價',DDprice)
            print('原價',price)
            print('學員',students)
            print('作者',teacher)
            print('時間',time_H,"時",time_M,"分")
            print('狀態',Tyep)
            ws.append([name,DDprice,price,students,teacher,Stype,time_H,time_M,Tyep])
            wb.save("友讀.xlsx")
            print("------------------------------------------------------------------------")
            sleep(1)
        print("請稍等翻頁中...")
        sleep(5)
    xx+=1
print("資料已爬完")
```
